# NLP INTRODUCTION
1. What is Computational Linguistics and relation with NLP?
>>Computational Linguistics is a field that deals with the use of computers to analyze and model human language. It focuses on linguistic rules and structures. Natural Language Processing (NLP) is an application of computational linguistics that enables machines to understand, interpret, and generate human language. NLP uses these linguistic models to perform real-world tasks such as translation, chatbots, and sentiment analysis.

2. Evolution of NLP
>> The evolution of NLP can be divided into stages:
Rule-based systems (1950s–70s): Based on grammar rules
Statistical methods (1980s–2000s): Used probability and data
Machine Learning era: Algorithms learned from data
Deep Learning era (present): Neural networks like BERT and GPT
Modern NLP focuses on context understanding and accuracy.

3. Three Use Cases of NLP
Answer:
Chatbots: Automate customer service interactions
Sentiment Analysis: Analyze user opinions from reviews
Machine Translation: Convert text from one language to another
These applications improve automation and decision-making.

4. What is Text Normalization?
>>Text normalization is the process of converting text into a standard format by removing noise such as punctuation, stopwords, and special characters. It includes lowercasing and cleaning data. It is important because it improves the quality of input data, leading to better performance of NLP models.

5. Stemming vs Lemmatization
>>Stemming: Removes suffixes (running → runn)
Lemmatization: Converts to base form using dictionary (running → run)
Stemming is faster but less accurate, while lemmatization is more accurate and meaningful.



```python
#6. Regex Email Extraction
import re

text = """Hello team, please contact us at support@xyz.com,
hr@xyz.com, john.doe@xyz.org, jenny_clarke126@mail.co.us,
partners@xyz.biz"""

emails = re.findall(r'\b[\w\.-]+@[\w\.-]+\.\w+\b', text)

print(emails)
```

```python
#7. Tokenization & Frequency (NLTK)
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Added based on error suggestion
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

text = """Natural Language Processing is a fascinating field."""

tokens = word_tokenize(text.lower())

fdist = FreqDist(tokens)

print(tokens)
print(fdist)
```

```python
#8. Proper Noun Detection (Corrected Code)

!pip install spacy
!python -m spacy download en_core_web_sm

import spacy

nlp = spacy.load("en_core_web_sm")

text = "John lives in New York and works at Google."

doc = nlp(text)

for token in doc:
    if token.pos_ == "PROPN":
        print(token.text)
```

```python
#9. Word2Vec using Gensim
!pip install gensim
from gensim.models import Word2Vec
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Adding this to resolve LookupError
from nltk.tokenize import word_tokenize

dataset = [
 "Natural language processing enables computers",
 "Word embeddings represent words",
 "Word2Vec is popular"
]

data = [word_tokenize(sentence.lower()) for sentence in dataset]

model = Word2Vec(sentences=data, vector_size=100, window=5, min_count=1)

print(model.wv['word'])
```
